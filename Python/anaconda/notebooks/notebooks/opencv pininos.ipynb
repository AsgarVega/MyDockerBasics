Empezamos importando las librerias correspondientes

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

definimos la funcion para la impresion de nuetras imagenes

```python
def printer(img,title): # impresion por defecto
    tmp=cv.cvtColor(img,cv.COLOR_BGR2RGB)
    plt.rcParams["figure.figsize"] = (20,20)
    plt.imshow(tmp)
    plt.title(title)
    plt.show()
```

cargamos la imagen con la que vamos a trabajar y la probamos

```python
cat_original = cv.imread('./imag/cats.jpg')
printer(cat_original,'gatito')
```

Tratamientos de color

```python
cat_gris = cv.cvtColor(cat_original, cv.COLOR_BGR2GRAY) # cv.Color_BGR2<type>
printer(cat_gris,'Gris')
```

Blur

```python
kernel_1D = 3
kernel_2D = (kernel_1D,kernel_1D) # tamaño minimo de (1,1) que es sin blur, tamaño maximo (9,9)
# viene a ser el tamaño del pincel (realmente es el tamaño de la matriz que procesara la imagen)
blur = cv.blur(cat_original, kernel_2D)

gauss = cv.GaussianBlur(cat_original, kernel_2D, 0)

median = cv.medianBlur(cat_original,kernel_1D)

bilateral = cv.bilateralFilter(cat_original,9,15,15)
```
