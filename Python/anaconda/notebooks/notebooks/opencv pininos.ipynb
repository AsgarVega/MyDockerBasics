*IMPORTANTE* no se usara opencv.imshow(...) porque en primera usa el mismo proceso que el kernel y puede dificultar el uso en notebooks de manera normal y en segunda por ser un entorno dockerizado, este entorno no cuenta con entorno grafico que mostrar, asi que se usará matplotlib.pyplot como visualizador, ademas se establecera una funcion personalizada para mostrar informacion ya que opencv usa el formato BGR y matplotlib usa RGB

--------------------------------------------------------------------

Empezamos importando las librerias correspondientes

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

definimos la funcion para la impresion de nuetras imagenes

```python
def printer(img,title):
    tmp=cv.cvtColor(img,cv.COLOR_BGR2RGB)
    plt.rcParams["figure.figsize"] = (10,20)
    plt.imshow(tmp)
    plt.title(title)
    plt.show()
```

cargamos la imagen con la que vamos a trabajar y la probamos

```python
cat_original = cv.imread('./imag/cat.jpg')
printer(cat_original,'gatito')
```
